### DETALHES TÉCNICOS DA BIBILITECA DE MACHINE LEARNING:
  - O Support Vector Machine (SVM) é um algoritmo de aprendizado de máquina supervisionado usado para classificação e regressão. Embora também digamos problemas de regressão, é mais adequado para classificação. 
  - O principal objetivo do algoritmo SVM é encontrar o hiperplano ótimo em um espaço N-dimensional que possa separar os pontos de dados em diferentes classes no espaço de recursos. 
  - O hiperplano tenta que a margem entre os pontos mais próximos de diferentes classes seja a máxima possível. A dimensão do hiperplano depende do número de feições. Se o número de feições de entrada for dois, então o hiperplano é apenas uma linha. 
  - Se o número de recursos de entrada for três, o hiperplano se tornará um plano 2-D. Torna-se difícil imaginar quando o número de recursos excede três. 
  https://www.geeksforgeeks.org/support-vector-machine-algorithm/

### Criação de uma função para detalhamento dos dados Null ( data null ) e Types
  - Criarei duas funções e analisarei qual melhor se encaixa para os processos

```python
# Função para mapear dos dados do dataset 

def summary_missing_values(data):
    missing_values = pd.DataFrame(df_raw.isna().sum()/df_raw.shape[0])
    missing_values.reset_index(drop=True, inplace=True)    
    missing_values.columns = ["Absoluto dados ausentes(%)"]
    missing_values["Column"] = pd.Series(df_raw.columns).values
    # Make index be the column "Absolute Missing (%)"
    missing_values.reset_index(drop=True, inplace=True)
    missing_values.set_index("Column", inplace=True)
    
    # Calculando os valores comparando com o tatal de dados por coluna e os dados ausentes quantitativos em percentual
    column_missing = []
    column_total   = []
    column_miss    = []
    column_type    = []
    
    for column in df_raw.columns:
        col_miss = df_raw[column].isnull().sum()
        total_values = df_raw[column].count()       
        column_miss.append(col_miss)
        column_total.append(total_values) 
        column_missing.append((col_miss/total_values))
        column_type.append(column_type)
        
    # Construindo o dataframe que receberá os valores
    missing_values["Perc. Abs. ausentes(%)"] = pd.Series(column_missing).values
    missing_values["Perc. da equivalência(%)"] = -(pd.Series(column_missing).values-1)
    missing_values["Vlres únicos"] = pd.Series(column_total).values
    missing_values["Valores abs ausentes"] = pd.Series(column_miss).values
    missing_values['Variáveis tipos'] = pd.DataFrame(df_raw.dtypes, columns=['column_type'])
    missing_values.sort_values(by="Absoluto dados ausentes(%)", ascending=False, inplace=True)
    missing_columns = list(missing_values.query("`Absoluto dados ausentes(%)` > 0").index)
    
    pd.options.display.float_format = '{:.2%}'.format
    return missing_values, missing_columns

print('\n\t\t\t\t\t\tMATRIZ DE MAPEAMENTO DOS DADOS\n'.center(100))

missing_values, missing_columns_names = summary_missing_values(df_raw)
display(missing_values)
```

```python
def statistic_summary(df_raw): 
    df_raw = df_raw[numerical_features]   
    # count_ = pd.DataFrame(df_raw.apply(np.size))
    # pct_ = pd.DataFrame(df_raw.apply(np.size)/len(numerical_features)*100)
    
    sum_ = pd.DataFrame(df_raw.apply(np.sum))
    mean_ = pd.DataFrame(df_raw.apply(np.mean))
    std_ = pd.DataFrame(df_raw.apply(np.std))
    median_ = pd.DataFrame(df_raw.apply(np.median))
    
    min_ = pd.DataFrame(df_raw.apply(min))
    max_ = pd.DataFrame(df_raw.apply(max))
    
    range_ = pd.DataFrame(df_raw.apply(lambda x: x.max() - x.min()))
    q1_ = pd.DataFrame(df_raw.apply(lambda x: np.percentile(x,25)))
    q3_ = pd.DataFrame(df_raw.apply(lambda x: np.percentile(x,75)))
    
    skewness_= pd.DataFrame(df_raw.apply(lambda x: x.skew()))
    kurtosis_ = pd.DataFrame(df_raw.apply(lambda x: x.kurtosis()))
    
    # summary = pd.concat([count_, pct_, mean_, std_, median_, sum_, min_, max_, range_, q1_,q3_],axis=1)
    summary = pd.concat([mean_, std_, median_, sum_, min_, max_, range_, q1_, q3_, skewness_, kurtosis_ ],axis=1)
    summary.columns = ['MEDIA','DESVIO','MEDIANA','SOMA','MIN','MAX', 'RANGE', '25%','75%', 'SKEW', 'KURTOSIS']

    cm = sns.light_palette('Blue', as_cmap = True)
    display(summary.style.format('{:.2f}').background_gradient(subset=(summary.index[2:],summary.columns[:]),cmap=cm))
```

### Criação de uma função para plotagens dos gráficos e cenarizações.
  - Para cada plot, gráfico criado, chamo essa function para tratar do refinamento dos gráficos
   - Parâmetros plt %matplotlib para as labels 
   - Bar: gráficos verticais
   - Barh: gráficos horizontais
___

```python
### Function 
'''
  Dbian Data Scientist 
  Função responsável por anotar os indicadores das barras no gráfico.
  Entrada: tipo do gráfico ('bar' ou 'barh')
'''
def autolabel(tipo):
  if tipo == 'bar':
    for bar in plot.patches:
      plot.annotate(format(bar.get_height()),
                xy = (bar.get_x() + bar.get_width() / 2, bar.get_height()),
                xytext=(0, 8),
                textcoords='offset points',
                ha='center', 
                va='center',
                size=11)
  elif tipo == 'barh':
    for bar in plot.patches:
      plot.annotate(format(bar.get_width()),
                  xy = (bar.get_width() + 0.3, bar.get_y() + 0.38),
                  xytext=(15, -2),
                  textcoords='offset points',
                  ha='center', 
                  va='center',
                  size=11)
    
# Função para retirar o contorno dos plots - gráficos. 
def gca():
    for posicao in ['right', 'top', 'bottom', 'left']: 
        plt.gca().spines[posicao].set_visible(False) 
    return None
```

```python
def modelo_matrix(pred_y,test_y):
     
    # Labels, title and ticks
    sns.set(font_scale=2.0) # Adjust to fit
    label_font = {'size':'15'}  # Adjust to fit
    ax.set_xlabel('Predicted labels', fontdict=label_font);
    ax.set_ylabel('Observed labels', fontdict=label_font);
    
    title_font = {'size':'15'}  # Adjust to fit
    ax.set_title('Confusion Matrix', fontdict=title_font);

    ax.tick_params(axis='both', which='major', labelsize=10)  # Adjust to fit
    ax.xaxis.set_ticklabels(['False', 'True']);
    ax.yaxis.set_ticklabels(['False', 'True']);
    
    cm=confusion_matrix(test_y,pred_y)
    class_label = ["High-risk", "Low-risk"]
    df_cm = pd.DataFrame(cm, index=class_label,columns=class_label)
    sns.heatmap(df_cm,annot=True, cmap='Pastel1',linewidths=2, fmt='d')
    plt.title('MATRIX DE CONFUSÃO',fontsize=15)
    plt.xlabel("Predicted")
    plt.ylabel("True")

    # Cálculo das métricas dos modelos implementados
    
    acc=accuracy_score(test_y,pred_y)
    mse=mean_squared_error(test_y,pred_y)
    precision, recall, fscore, train_support = score(test_y, pred_y, pos_label=1, average='binary')
    
    print(f'Tempo de execução do modelo: {round((model_Log_time),5)} seconds\n')
    print(f'\nMÉTRICAS DO MODELO\n')
    print('Precision: {} \nRecall: {} \nF1-Score: {} \nAccuracy: {} %\nMean Square Error: {}'.format(
        round(precision, 3), round(recall, 3), round(fscore,3), round((acc*100),3), round((mse),3)))
    
    plt.tight_layout()
```

### Aplicarei uma function (função)
  - Retorno esperado: Checar o modelo que tem o melhor desempenho usando a seguinte função checkmodel():

```python
def checkmodel():
    mse_dict={}  # Root mean square dictionary
    acc_dict={}  # Accuracy dictionary

    for n in range(25,400,25):
        train_x, test_x, train_y, test_y = train_test_split(X, Y, test_size=0.2,random_state=65)
        model_mlp = MLPClassifier(random_state=48,hidden_layer_sizes=(150,100,50), max_iter=n,activation = 'relu',solver='adam')
        model_mlp.fit(train_x, train_y)
        pred_y=model_mlp.predict(test_x)
        acc=accuracy_score(test_y,pred_y)
        mse=mean_squared_error(test_y,pred_y)
        mse_dict.update({n:round(mse,3)})
        acc_dict.update({n:round((acc*100),3)})

    # Mean Square Error
    lowest=min(mse_dict.values())
    res = [key for key in mse_dict if mse_dict[key] == lowest]
    mse_list=mse_dict.items()
    k,v = zip(*mse_list) 
    print(f'\nMELHORES MÉTRICAS\n')
    print("RMSE mais baixo é em: {} for n: {} ".format(round((lowest),3),res))

    # Plot RMSE values
    plt.figure(figsize=(12,6))
    sns.set(font_scale=1.0) # Adjust to fit
    # plt.scatter(res,lowest,color="red",lw=5)
    plt.plot(k,v)
    plt.xlabel("Random State")
    plt.ylabel("RMSE")
    plt.grid(False)
    gca()
    plt.box( False )
    
    plt.tight_layout()

    # Accuracy
    highest=max(acc_dict.values())
    res1= [key for key in acc_dict if acc_dict[key] == highest]
    acc_list=acc_dict.items()
    k1,v1=zip(*acc_list)
    print("Accuracy is highest at {} % for n: {} ".format(highest,res1))
    # Plot Accuracy values
    sns.set(font_scale=1.0) # Adjust to fit
    plt.figure(figsize=(12,6))
    # plt.scatter(res1,highest,color="red",lw=5)
    plt.plot(k1,v1)
    plt.xlabel("Random State")
    plt.ylabel("Accuracy")
    plt.grid(False)
    gca()
    plt.box( False )
    
    plt.tight_layout()
    
checkmodel()
```
